```python
from matplotlib import pyplot as plt
```

```python
# USER INPUT
instrument = 'call' # Enter 'call' or 'put'
strike_price = 45
size = 1
initial_price = 2.88
```

```python
if instrument not in ['call', 'put']:
    print('Invalid input')
```

```python
qty = size*100
startpt = int(strike_price-strike_price/2)
endpt = int(strike_price+strike_price/2)
underlying_price = list(range(startpt,endpt))
cashflow = []
if instrument == 'call':
    for x in underlying_price:
        cashflow.append(((max(x - strike_price,0))-initial_price)*qty)
elif instrument == 'put':
    for x in underlying_price:
        cashflow.append(((max(strike_price - x,0))-initial_price)*qty)

if len(cashflow)==len(underlying_price):
    plt.plot(underlying_price,cashflow)
    plt.title('%s option payout diagram' % (instrument))
    plt.xlabel('Current price')
    plt.ylabel('P/L')
    plt.grid(True, linestyle='-.')
    plt.show()
else:
    print('Something went wrong. Confirm user input for instrument type')

```
